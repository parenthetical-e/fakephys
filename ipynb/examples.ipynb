```python
from fakepark import model, noise
from bokeh.plotting import figure
from bokeh.io import show, output_notebook
from bokeh.layouts import column, row
output_notebook()

import numpy as np
from scipy.signal import welch
```

```python
# GLOBAL params
t = 20
times = model.create_times(t, 1e-3)
```

# Examples


Examples timecourses for each of the models.

## IE (A wilson cowan system).

Frequency band selection:
 - Alpha {tau_e = N, tau_i = N}
 

```python
x = model.ie(t, P=4)

# -
freqs, psd = welch(x, int(1 / 1e-3), scaling='density', nperseg=1000/2)
m = np.logical_and(freqs > 2, freqs < 100)
freqs = freqs[m]
psd = psd[m]

# -
p1 = figure(plot_width=600, plot_height=300)
p1.line(x=times, y=x, color="black")
p1.xaxis.axis_label = "Time (s)"
p1.yaxis.axis_label = "Activity"

p2 = figure(plot_width=300, plot_height=300)
p2.line(x=freqs, y=psd, color="black")
p2.xaxis.axis_label = "Freq (Hz)"
p2.yaxis.axis_label = "Power (AU)"
show(row(p1, p2))
```

## Ramping input drive

```python
x = model.slidie(t, 1, 5, tau_e=3e-3, tau_i=80e-3, sigma=0)

# -
freqs, psd = welch(x, int(1 / 1e-3), scaling='density', nperseg=1000/2)
m = np.logical_and(freqs > 2, freqs < 100)
freqs = freqs[m]
psd = psd[m]

# -
p1 = figure(plot_width=600, plot_height=300)
p1.line(x=times, y=x, color="black")
p1.xaxis.axis_label = "Time (s)"
p1.yaxis.axis_label = "Activity"

p2 = figure(plot_width=300, plot_height=300)
p2.line(x=freqs, y=psd, color="black")
p2.xaxis.axis_label = "Freq (Hz)"
p2.yaxis.axis_label = "Power (AU)"
show(row(p1, p2))
```
